```python
#Q1. What is the maximum number of matches played by an individual player in a season?

matches_by_player = df['Mat'].max()
print('Maximum matches played is:', matches_by_player)
```

```python
#Q2. What is the highest average of a player in a season?

highest_average_by_player = df['Avg'].max()
a=df["Avg"].idxmax()
value=df.loc[a, "Player"]
print('Highest average of a player is:', value, highest_average_by_player)
```

```python
#Q3. What is the average number of matches played by a player in a season?

avg_matches_by_player = df['Mat'].mean()
x=np.mean(df['Mat'])
y=int(x)
print('Average matches played is:', y)
```

```python
#Q4. Please check the correlation between all the features.

cor=df.corr()
plt.figure(figsize=(15,10))
sns.heatmap(cor,cmap="coolwarm", annot=True, fmt=".2f", square=True)
plt.title("correlation matrix")
plt.show()
```

```python
#Q5. Check the list of players who has an average greater than 50 as well strike rate above 120.

high_avg_sr = df.loc[(df['Avg'] > 50) & (df['SR'] > 120), 'Player'].tolist()
high_avg_sr
```

```python
#Q6. Please check the list of players who has an average greater than 40 and balls faced above 100.

high_avg_bf = df.loc[(df['Avg'] > 40) & (df['BF'] > 100), 'Player'].tolist()
high_avg_bf
```

```python
#Q7. Please check the list of players who scored more than 300 runs with an average greater than 45 & strike rate above 130.

runs_avg_sr = df.loc[(df['Runs'] > 300) & (df['Avg'] > 45) & (df['SR'] > 130), 'Player'].tolist()
runs_avg_sr
```

```python
#Q8. Players who scroed alteast one century in this season.

century = df.loc[(df['100'] >= 1), 'Player'].tolist()
century
```

```python
#Q9. Players who scored atleast 4 half centuries in this season.

half_century = df.loc[(df['50'] >= 4), 'Player'].tolist()
half_century
```

```python
#Q10. Check the list of players who hit more than 45 boundARIES and more than 10 sixes in this season.

boundries = df.loc[(df['4s'] > 45) & (df['6s'] > 10), 'Player'].tolist()
boundries
```

```python
#Q11. Plot a histogram of number of matches played in a season by players.

plt.hist(df['Mat'], bins=20)
plt.xlabel('Matched Played')
plt.ylabel('Players')
plt.title('Histogram of Number of matches played in a season by players')
plt.show()
```

```python
#Q12. Plot the histogram of balls faced by players.

plt.hist(df['BF'], bins=20)
plt.xlabel('Balls Faced')
plt.ylabel('Players')
plt.title('Histogram of Number of balls faced by players')
plt.show()
```

```python
#Q13. Top 30 players with most runs in a season v/s matches played.

runs_df = df.groupby('Player')['Mat','Runs'].sum()

top_30_players = runs_df.sort_values(by='Runs', ascending=False).head(30)

print(top_30_players)
```

```python
#Q14. Bar chart for top 20 players with best strike rate in a season.

top_20_players = df.sort_values(by='SR', ascending=False).head(20)

plt.figure(figsize=(12,8))
plt.bar(top_20_players['Player'], top_20_players['SR'], color='blue')
plt.xticks(rotation=90)
plt.xlabel('Player')
plt.ylabel('Strike Rate')
plt.title('Top 20 Players with Best Strike Rate in a Season')
plt.show()
```

```python
#Q15. Top 30 players with highest number of sixes.

highest_6s = df.groupby('Player')['Mat','6s'].sum()

top_30_players = highest_6s.sort_values(by='6s', ascending=False).head(30)

print(top_30_players)
```

```python
#Q16. Scatter plot of runs scored by a player v/s balls faced in a season.

plt.scatter(df['Runs'], df['BF'])
sns.regplot(x=df['Runs'], y=df['BF'])
plt.xlabel('Runs Scored')
plt.ylabel('Balls Faced')
plt.title('Scatter plot of runs scored by a player v/s balls faced in a season')
plt.show()
```

```python
#Q17. KDE plot for averages on a season.

sns.kdeplot(data=df, x='Avg')
plt.title('KDE plot for averages on a season.')
```
